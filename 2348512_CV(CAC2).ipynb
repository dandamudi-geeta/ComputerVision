**Affine transformation**

**We use WarpAffline to define the afflinetransformation**

```python
rows,cols, ch=img.shape
pts1 = np.float32([[50, 50], [200, 50], [50, 200]])
pts2 = np.float32([[10, 100], [200, 50], [100, 250]])
M = cv2.getAffineTransform(pts1, pts2)
```

```python
# Apply the affine transform
dst = cv2.warpAffine(img, M, (cols, rows))
```

```python
#displaying the original image and output transformed image
# Display the original and transformed images
plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Input')
plt.subplot(122), plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)), plt.title('Output')
plt.show()
```

**Translation**

```python
rows,cols,ch=img.shape
```

```python
M=np.float32([[1,0,100],[0,1,50]])
```

```python
#transform image

dst = cv2.warpAffine(img, M, (cols, rows))
```

```python
# Display the original and transformed images
plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Input')
plt.subplot(122), plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB)), plt.title('Output')
plt.show()
```

**Scaling**

```python
# Scale the image
s_x, s_y = 0.75, 0.75
mat_scale = np.array([[s_x, 0, 0], [0, s_y, 0], [0, 0, 1]])
img_scale = ndi.affine_transform(img, mat_scale, order=1)
plt.imshow(img_scale, cmap="gray")
plt.title('Scaling')
plt.show()
```

**Rotating The Image**

```python
from skimage.io import imread
from skimage.color import rgb2gray
```

```python

# Read and convert the image to gray

img = rgb2gray(imread('/content/Gold Fish.jpg'))
w, h = img.shape
```

```python
w, h = img.shape
```

```python
# Rotate the image
theta = np.pi / 6
mat_rotate = np.array([[1, 0, w / 2], [0, 1, h / 2], [0, 0, 1]]) @ \
             np.array([[np.cos(theta), np.sin(theta), 0], [np.sin(theta), -np.cos(theta), 0], [0, 0, 1]]) @ \
             np.array([[1, 0, -w / 2], [0, 1, -h / 2], [0, 0, 1]])
img_rotate = ndi.affine_transform(img, mat_rotate, order=1)
plt.imshow(img_rotate, cmap="gray")
plt.title('Rotation')
plt.show()
```

**Average Filter**

```python
#average filter technique
from google.colab.patches import cv2_imshow

#load the image
img=cv2.imread("/content/Gold Fish.jpg")
```

```python
cv2_imshow(img)
```

```python
# Define the kernel for the average filter
kernel = np.ones((5, 5), np.float32) / 25  # 5x5 kernel for average filter

# Apply the average filter
average_filtered_image = cv2.filter2D(img, -1, kernel)
```

```python
cv2_imshow(average_filtered_image)
```
