```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import convolve
from scipy import ndimage
```

```python
imgPath = '/content/Gold Fish.jpg'
img = cv2.imread(imgPath,cv2.IMREAD_GRAYSCALE)
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

**Defining Kernals**

```python
# Identity Kernel
identityKernel = np.array([[0,0,0],
                           [0,1,0],
                           [0,0,0]])
```

```python
# EdgeDetection kernel
edgeDetectionKernel = np.array([[-1,-1,-1],
                                [-1,8,-1],
                                [-1,-1,-1]])
```

```python
#BoxBoxKernel
boxBlurKernel = np.ones((3,3),dtype=np.float32) / 9.0
```

```python
# Sharpen Kernel
sharpenKernel = np.array([[0,-1,0],
                          [-1,5,-1],
                          [0,-1,0]])
```

```python
#GaussianKernel

gaussianBlurKernel = cv2.getGaussianKernel(5,0)
```

**Applying filter 2D**

```python
identityKernelResult = cv2.filter2D(img,-1,identityKernel)
edgeDetectionKernelResult = cv2.filter2D(img,-1,edgeDetectionKernel)
sharpenKernelResult = cv2.filter2D(img,-1,sharpenKernel)
boxBlurKernelResult = cv2.filter2D(img,-1,boxBlurKernel)
gaussianBlurKernelResult = cv2.GaussianBlur(img,(5,5),0)
```

```python
#Visualization

plt.figure(figsize=(10,6))

plt.subplot(2,3,1), plt.imshow(img,cmap='gray')
plt.title('Original Image')
```

```python
plt.subplot(2,3,2), plt.imshow(identityKernelResult,cmap='gray')
plt.title('Identity Kernel')
```

```python
plt.subplot(2,3,3), plt.imshow(edgeDetectionKernelResult,cmap='gray')
plt.title('EdgeDetection Kernel')
```

```python
plt.subplot(2,3,4), plt.imshow(sharpenKernelResult,cmap='gray')
plt.title('Sharpen Kernel')
```

```python
plt.subplot(2,3,5), plt.imshow(boxBlurKernelResult,cmap='gray')
plt.title('BoxBlur Kernel')
```

```python
plt.subplot(2,3,6), plt.imshow(gaussianBlurKernelResult,cmap='gray')
plt.title('GaussianBlurKernel')
```

**Applying Convolve function in SCIPY library and use the following mode reflect, constant, nearest, mirror, wrap'**

```python
image_path = '/content/Gold Fish.jpg'
img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
Kernel = ([[1, 1, 1],
          [1, -7, 1],
          [1, 1, 1]])
```

```python
#Using different modes in scipy library

modes = ['reflect','constant','nearest','mirror','wrap']
plt.figure(figsize=(15,8))

for i,mode in enumerate(modes,1):
  conv = convolve(img,Kernel,mode=mode,cval=0.0)
  plt.subplot(2,3,i), plt.imshow(conv,cmap='gray')
  plt.title(f'Mode:{mode}')
  plt.axis('off')


plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')

plt.show()

```

**Conclusion**

Performed correlation and convolution operations by applying different kernels, and performed convolution using convolve function in scipy library by using different modes.

When performing convolution using scipy library by uitilizing different modes, I'hve not witnessed any specific changes for each mode.

Reflect: This mode reflects the image at the boundaries, creating a mirror-like effect.

Constant: In this mode, the values of the boundaries are extended from the edge pixels of the image.

Nearest: This mode simply uses the value of the nearest pixel for the boundary pixels.

Mirror: Similar to the reflect mode, but the image is mirrored with alternate sections flipped.

Wrap: This mode essentially tiles the image, wrapping it at the boundaries.
